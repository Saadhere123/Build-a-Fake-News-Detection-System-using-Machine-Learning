```python
import joblib
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
model = joblib.load('/content/model.pk1')
vectorizer = joblib.load('/content/vectorizer.pk1')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(text):
  text = text.lower()
  text = re.sub(r'[^a-zA-Z]', ' ', text)
  words = text.split()
  words = [lemmatizer.lemmatize(w) for w in words if w not in stop_words]
  return " ".join(words)
```

```python
news = input("Enter News Text : ")

clean_text = preprocess(news)
vector = vectorizer.transform([clean_text])

prediction = model.predict(vector)[0]
probability = model.predict_proba(vector).max()

print("\nPrediction : ", "Fake News" if prediction == "Fake" else "Real News")
print("Confidence : ", round(probability * 100, 2), "%")
```
